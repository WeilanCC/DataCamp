```python
import pandas as pd
from sklearn.utils import shuffle
```

```python
filename = 'kaggle_data/fraud_detection.csv'
data = pd.read_csv(filename)
data.head()
```

```python
data.insert(0, 'id', range(len(data)))
data.head()
```

```python
data = shuffle(data)
data.head()
```

```python
size = len(data)
TARGET = 'isFraud'
data_train = data.iloc[:int(0.8 * size)]
data_test = data.iloc[int(0.8 * size):]
```

```python
data_sample_submission = data_test[['id', TARGET]].copy()
data_test = data_test.drop(TARGET, axis=1)
```

```python
data_train.head()
```

```python
data_test.head()
```

```python
data_sample_submission.head()
```

```python
data_train.to_csv('kaggle_data/train.csv', index=False)
data_test.to_csv('kaggle_data/test.csv', index=False)
data_sample_submission.to_csv('kaggle_data/sample_submission.csv', index=False)
```
